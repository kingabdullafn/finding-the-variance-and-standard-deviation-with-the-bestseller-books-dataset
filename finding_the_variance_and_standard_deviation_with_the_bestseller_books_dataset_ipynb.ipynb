```python
df = pd.read_csv("/content/Bestsellers with categories (1).csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.fillna(0)
```

```python
var_rating =np.var(df['User Rating'])
print("The variance of the User rating is ", var_rating)
```

```python
std_rating = np.std(df["User Rating"])
print("The standard deviation of the user rating is ",std_rating)
```

```python
var_price = np.var(df['Price'])
print("The variance of the price is ",var_price)
```

```python
std_price = np.std(df['Price'])
print("The standard deviation of the price is ", std_price)
```
